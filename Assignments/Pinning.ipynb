# 못 박기 (Pinning)

<br>

### 문제
- 아래 그림 처럼 n개의 막대가 입력으로 주어질 때, n개 막대를 최소 개수의 못(pin)으로 꽂으려 한다.
- 모든 막대는 최소한 하나 이상의 못이 박혀 있어야 한다.
- 이 때, 필요한 핀의 최소 개수는 몇 개인지 구하는 Greedy 알고리즘을 구현하라.

### 알고리즘

<br>

- 1. 막대의 개수 n과, 반복문을 통해 각 막대의 왼쪽 끝 점(a)와 오른쪽 끝 점(b)를 입력 받는다.
- 2. 입력된 a와 b를 튜플로 묶어 sticks 리스트에 추가한다.
- 3. sticks 리스트를 끝점(x[1])을 기준으로 정렬한다. (끝점이 동일한 경우에는 시작점(x[0])을 기준으로 정렬된다.)
- 4. cnt_pins로 핀이 꽂힌 막대 개수를 저장하고, pins 리스트 꽂힌 핀의 위치를 확인한다.
- 5. for문을 돌며 sticks 리스트의 각 막대에 대해, 현재 막대의 시작점(a)과 꽂힌 핀의 위치(=pins 리스트의 마지막 요소)를 비교한다.
    - 5-2. 만약 현재 막대의 시작점이 핀이 꽂힌 곳의 위치보다 크면, 새로운 핀을 꽂을 수 있다. 따라서 pins 리스트에 현재 막대의 끝점(b)을 추가하고 cnt_pins를 1 증가시킨다.
    - 5-3. 그리고 cnt_pins 변수를 반환한다.

<br>

### 수행시간

<br>

- 전체 수행 시간 = O(n log n)
    - 각 막대의 양 끝 점을 입력받는 부분 = O(n)
    - 막대를 정렬하는 부분 = O(n log n)
    - 막대의 수만큼 반복문을 실행하는 부분 = O(n)

<br>


```python
def pinning():
    n = int(input())
    sticks = []

    for _ in range(n):
        a, b = map(int, input().split())
        sticks.append((a, b))

    sticks.sort(key=lambda x: (x[1], x[0]))
    cnt_pins = 0
    pins = []

    for i in range(len(sticks)):
        a, b = sticks[i]
        if i == 0 or a > pins[-1]:
            pins.append(b)
            cnt_pins += 1

    return cnt_pins

print(pinning())
```
