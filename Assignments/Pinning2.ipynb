### 문제

- 아래 그림에서는 붉은색 위치에 핀을 꽂으면 8개의 막대까지 꽂을 수 있다.
- 못 하나로 관통할 수 있는 막대의 최대 개수를 계산하는 그리디 알고리즘을 구현하라.
- 주의 : 못이 막대 끝을 통과하더라도 꽂은 것으로 간주한다.

### 알고리즘

<br>


- 1. 막대 개수 n을 입력 받고, n번 반복하면서 왼쪽 끝 점a와 오른쪽 끝 점 b를 입력 받아 튜플 형태로 sticks 리스트에 추가한다.
- 2. max_num은 막대들의 끝 지점 중 가장 큰 값을 저장하기 위한 변수이다. 현재 b와 비교하여 더 큰 값을 저장한다.
- 3. cnt 는 각 지점마다 겹치는 막대의 개수를 세기 위한 리스트이다.
    - cnt에 시작 지점 a에서는 1을 더하고, 끝 지점 b에서는 다음 인덱스에 -1을 더한다. 
- 4. max_pins는 가장 많이 겹치는 막대의 개수를 저장하는 변수이고, current_pins는 현재 까지의 겹치는 막대의 개수를 저장하는 변수이다.
    - for문을 돌면서  current_pins에 cnt의 해당 인덱스 값을 더하고, max_pins와 비교하여 더 큰 값을 max_pins에 저장한 후 반환한다.

<br>

### 수행시간

<br>

-  총 수행 시간 = O(n)
    - 입력 받는 시간 = O(n)
    - 위의 3번 반복문에서의 수행 시간 = O(n)
    - 위의 4번 반복문에서의 수행 시간 = O(n)
    
<br>

```python
def find_max():
    n = int(input())
    sticks = []
    max_num = 0; max_pins = 0; current_pins = 0

    for _ in range(n):
        a, b = map(int, input().split())
        sticks.append((a, b))
        max_num = max(max_num, b)

    cnt = [0] * (max_num+2)

    for stick in sticks:
        a = stick[0]; b = stick[1]
        cnt[a] += 1
        cnt[b + 1] += -1

    for i in range(len(cnt)):
        current_pins += cnt[i]
        max_pins = max(current_pins, max_pins)

    return max_pins

print(find_max())
```
